```python
import pandas as pd # 导入pandas模块

source_file = "数据.xlsx" # 定义Excel文件路径

# 读取Excel数据
df = pd.read_excel(
    
    source_file, # Excel文件路径
    
    sheet_name='序时账', #获取'序时账'工作表的数据
    
    header=0, #取第1行的值作为列索引
    
    index_col=0, # 把第1列数据设置为行索引
    # 只取日期、凭证号、科目编号、科目名称这四列数据
    
    usecols=['日期','凭证号','科目编号','科目名称'], 
    # 把凭证号、科目编号这两列的数据类型修改为字符串
    
    dtype={'凭证号' : str,'科目编号' : str}, 
    skipfooter=4 # 跳过最后4行的数据,从倒数第5行开始向上取数据
)
df
```
